```python
import pandas as pd

datasets = {
    "fund_master":"01_fund_master.csv",
    "nav_history":"02_nav_history.csv",
    "aum":"03_aum_by_fund_house.csv",
    "sip":"04_monthly_sip_inflows.csv",
    "category":"05_category_inflows.csv",
    "folios":"06_industry_folio_count.csv",
    "performance":"07_scheme_performance.csv",
    "transactions":"08_investor_transactions.csv",
    "holdings":"09_portfolio_holdings.csv",
    "benchmark":"10_benchmark_indices.csv"
}

dfs = {}

for name,file in datasets.items():
    df = pd.read_csv(file)
    dfs[name] = df

    print("="*80)
    print(name.upper())
    print("\nShape:")
    print(df.shape)

    print("\nDtypes:")
    print(df.dtypes)

    print("\nHead:")
    display(df.head())
```

```python
for name,df in dfs.items():

    print("\n"+"="*80)
    print(name)

    print("Rows:",len(df))

    print("Missing Values:")
    print(df.isnull().sum().sum())

    print("Duplicate Rows:")
    print(df.duplicated().sum())
```

```python
# ==========================
# FUND MASTER EXPLORATION
# ==========================

fund_master = dfs["fund_master"]

print("="*80)
print("FUND MASTER EXPLORATION")
print("="*80)

print("\nColumns Available:")
for col in fund_master.columns:
    print("-", col)

# Fund Houses
if "fund_house" in fund_master.columns:
    print("\n" + "="*40)
    print("UNIQUE FUND HOUSES")
    print("="*40)
    print(fund_master["fund_house"].unique())
    print(f"\nTotal Fund Houses: {fund_master['fund_house'].nunique()}")

# Categories
if "category" in fund_master.columns:
    print("\n" + "="*40)
    print("CATEGORIES")
    print("="*40)
    print(fund_master["category"].unique())

# Find Sub-category column automatically
subcategory_cols = [
    "subcategory",
    "sub_category",
    "scheme_sub_category",
    "sub_category_name"
]

found_subcategory = False

for col in subcategory_cols:
    if col in fund_master.columns:
        print("\n" + "="*40)
        print("SUB CATEGORIES")
        print("="*40)
        print(fund_master[col].unique())
        found_subcategory = True
        break

if not found_subcategory:
    print("\nNo sub-category column found.")

# Find Risk column automatically
risk_cols = [
    "risk_grade",
    "risk_level",
    "risk",
    "riskometer"
]

found_risk = False

for col in risk_cols:
    if col in fund_master.columns:
        print("\n" + "="*40)
        print("RISK DISTRIBUTION")
        print("="*40)
        print(fund_master[col].value_counts())
        found_risk = True
        break

if not found_risk:
    print("\nNo risk column found.")
```

```python
# ==========================================
# CELL 6 - AMFI CODE VALIDATION (AUTO-DETECT)
# ==========================================

fund_master = dfs["fund_master"]
nav_history = dfs["nav_history"]

print("Fund Master Columns:")
print(fund_master.columns.tolist())

print("\nNAV History Columns:")
print(nav_history.columns.tolist())

# Possible code column names
possible_code_cols = [
    "scheme_code",
    "amfi_code",
    "amfi_scheme_code",
    "schemecode",
    "schemeid",
    "scheme_id",
    "fund_code",
    "code"
]

master_code_col = None
nav_code_col = None

# Find code column in fund_master
for col in possible_code_cols:
    if col in fund_master.columns:
        master_code_col = col
        break

# Find code column in nav_history
for col in possible_code_cols:
    if col in nav_history.columns:
        nav_code_col = col
        break

# If not found, use first column that contains "code"
if master_code_col is None:
    for col in fund_master.columns:
        if "code" in col.lower():
            master_code_col = col
            break

if nav_code_col is None:
    for col in nav_history.columns:
        if "code" in col.lower():
            nav_code_col = col
            break

print("\nDetected Columns")
print("Fund Master Code Column:", master_code_col)
print("NAV History Code Column:", nav_code_col)

if master_code_col is None or nav_code_col is None:
    print("\n❌ Could not automatically find scheme code columns.")
    print("Please share the column names shown above.")
else:

    master_codes = set(fund_master[master_code_col].astype(str))
    nav_codes = set(nav_history[nav_code_col].astype(str))

    missing_codes = master_codes - nav_codes

    print("\n" + "="*50)
    print("AMFI CODE VALIDATION")
    print("="*50)

    print("Total Schemes in Fund Master:", len(master_codes))
    print("Total Schemes in NAV History:", len(nav_codes))
    print("Missing Codes:", len(missing_codes))

    if len(missing_codes) > 0:
        print("\nSample Missing Codes:")
        print(list(missing_codes)[:10])
    else:
        print("\n✅ All scheme codes found in NAV history.")
```

```python
import requests
import pandas as pd

scheme_code = 125497

url = f"https://api.mfapi.in/mf/{scheme_code}"

response = requests.get(url)

print("Status Code:", response.status_code)

data = response.json()

print("\nScheme Name:")
print(data["meta"]["scheme_name"])

nav_df = pd.DataFrame(data["data"])

print("\nLatest NAV Records:")
display(nav_df.head())

# Save to Colab
nav_df.to_csv("HDFC_Top100_NAV.csv", index=False)

print("\nSaved as HDFC_Top100_NAV.csv")
```

```python
import requests
import pandas as pd

schemes = {
    "HDFC_Top100": 125497,
    "SBI_Bluechip": 119551,
    "ICICI_Bluechip": 120503,
    "Nippon_LargeCap": 118632,
    "Axis_Bluechip": 119092,
    "Kotak_Bluechip": 120841
}

for name, code in schemes.items():

    try:
        url = f"https://api.mfapi.in/mf/{code}"

        response = requests.get(url)
        data = response.json()

        df = pd.DataFrame(data["data"])

        filename = f"{name}_NAV.csv"

        df.to_csv(filename, index=False)

        print(f"✅ Saved: {filename}")

    except Exception as e:
        print(f"❌ Error fetching {name}")
        print(e)
```
